```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

Project dedicated to check if education level has influence on the healthy life expectancy. Based on the data from Eurostat database. 
First step: examine correlation between education level, BMI and fruit&vegetables intake.
Second step: examine additional factors: tobbaco, alcohol and PKB per capita.

```python
bmi= pd.read_csv('bmi/hlth_ehis_bm1e_1_Data.csv')
```

```python
bmi['BMI'].unique()
```

```python
bmi['GEO'].unique()
```

```python
bmi.describe(include = ['O'])
```

First conclusions: data frame contains 1224 records. There are 3 BMI values: Underweight, Normal and Overweigth. Data covers 34 countries (including collected information for the EU), 3 education levels, provides differentiation between man and women, but not the age. Therefore the column: AGE can also be removed.

bmi data cleaning: 
- remove columns: UNIT, AGE, Flag and Footnotes, as they contain information not relevant for the analysis
- check for missing data and either remove or fill. No missing values except column: Flag and Footnotes.
- as the analysis will be done per country, data for the whole EU can be removed
- remove data from 2014

```python
sns.heatmap(bmi.isnull())
```

```python
bmi.drop('UNIT', axis=1, inplace=True)
```

```python
bmi.drop('Flag and Footnotes', axis = 1, inplace=True)
```

```python
bmi.drop('AGE', axis = 1, inplace = True)
```

```python
countries = bmi['GEO'].unique().tolist()[2:]
```

```python
bmi_single_countries = bmi[bmi['GEO'].isin(countries)].copy()
bmi_single_countries
```

```python
bmi_2019 = bmi_single_countries[bmi_single_countries['TIME'] == 2019].copy()
bmi_2019
```

```python
def change_to_number(x):
    if x == ':':
        return 0
    else:
        return float(x)

bmi_2019['Value'] = bmi_2019['Value'].apply(lambda x: change_to_number(x))
```

```python
education_descriptions = bmi_2019['ISCED11'].unique().tolist()
education_simplified = ['lower', 'middle', 'upper']
education_dict = {name:number for name,number in zip(education_descriptions, education_simplified)}
education_dict
```

```python
bmi_2019['education_level'] = bmi_2019['ISCED11'].map(education_dict)
```

```python
bmi_2019
```

```python
pivot_bmi = bmi_2019.pivot_table(values = 'Value', index = 'education_level', columns = 'BMI').round(2)
```

```python
pivot_bmi
```

```python
pivot_bmi.plot(kind = 'bar', stacked=True, title = 'BMI vs education level')
```

```python
pivot_bmi_M_F = bmi_2019.pivot_table(values = 'Value', index = 'education_level', columns = ['BMI', 'SEX']).round(2)
```

```python
pivot_bmi_M_F
```

```python
pivot_bmi_M_F['Overweight'].plot.bar()
```

Conclusion: Based on data from all EU countries, in the group with highest education the mean % of overweight citizens is > 6 %points lower than in the groups with education levels 0-2 and 3-4. In the group with lowest education level the % of underweight citizens is significantly higher than in the other groups. When the split into males and females is taken into consideration, it becomes apparent that the % of underweight males decreases with the increasing education level, while in the females group it is between 4-5% regardless of the education. In the group with lowest education the % of overweight females and males is almost equal (52.17 and 53.71%, respectively), while in the 2 other groups there is a vast difference depending on sex: there is 15 to 20% points less overweight females than males.

```python
fv_consumption = pd.read_csv('fruit_vegetables/hlth_ehis_fv3e_1_Data.csv')
```

```python
fv_consumption
```

```python
fv_consumption['Value'] = fv_consumption['Value'].apply(lambda x: change_to_number(x))
```

```python
fv_consumption.describe(include = ['O'])
```

```python
fv_consumption['N_PORTION'].unique()
```

```python
sns.heatmap(fv_consumption.isnull())
```

fv_consumption data cleaning:

- remove columns: UNIT, AGE,  Flag and Footnotes, as they contain information not relevant for the analysis
- check for missing data and either remove or fill. No missing values except column: Flag and Footnotes.
- as the analysis will be done per country, data for the whole EU can be removed
- remove data from 2014
- create column with education level as in bmi Data Frame.

```python
fv_consumption.drop('UNIT', axis=1, inplace=True)
```

```python
fv_consumption.drop('AGE', axis=1, inplace=True)
```

```python
fv_consumption.drop('Flag and Footnotes', axis=1, inplace=True)
```

```python
fv_consumption_single_countries = fv_consumption[fv_consumption['GEO'].isin(countries)].copy()
fv_consumption_single_countries
```

```python
fv_consumption_2019 = fv_consumption_single_countries[fv_consumption_single_countries['TIME'] == 2019].copy()
fv_consumption_2019
```

```python
fv_consumption_2019['education_level'] = fv_consumption_2019['ISCED11'].map(education_dict)
fv_consumption_2019
```

```python
portions = fv_consumption_2019['N_PORTION'].unique().tolist()
portions_new = ['0', '1-4', '5 or more']
portions_dict = dict(zip(portions, portions_new))
fv_consumption_2019['portions'] = fv_consumption_2019['N_PORTION'].map(portions_dict)
```

```python
fv_consumption_2019
```

```python
pivot_fv_cons = fv_consumption_2019.pivot_table(values = 'Value', index = 'education_level', columns = 'portions').round(2)
pivot_fv_cons
```

```python
pivot_fv_cons.plot.bar(stacked = True, title = 'Consumption of fruit and vegetables vs education level')
```

```python
pivot_fv_cons_M_F = fv_consumption_2019.pivot_table(values = 'Value', index = 'education_level', columns = ['portions', 'SEX']).round(2)
pivot_fv_cons_M_F
```

```python
pivot_fv_cons_M_F['5 or more'].plot.bar()
```

Conclusions: consumption of fruit and vegetables is clearly higher in the group with higher education level. Among EU citizens with lower education only 8.74% consume 5 or more portions per day. Even in the best educated group almost 30% of citizens don't consume fruit and vegetables at all. In all groups females tend to eat more fruit and vegetables than men. Further promotion of fruit and vegetables intake needs to take place, as these are the best sources of vitamins, minerals, dietary fiber and many other nutrients. Intake of 5 portions of diverse fruit and vegetables every day is known to have positive effect on health, prevent cancer and cardiovalcular disease.
